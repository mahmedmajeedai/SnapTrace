### Dataset sources  
Bad Quality Images (Wider): `http://shuoyang1213.me/WIDERFACE/`  
Good Quality Images (FFHQ): `https://github.com/NVlabs/ffhq-dataset`  

### Data Augmentation
We are having 2000 good images, and 1500 bad images, that causes data imbalance. To avoid this we will add 500 images **augmentation** in bad images.  

```python
import os
import random
from PIL import Image
from torchvision import transforms

TARGET         = 500

os.makedirs(FAIL_IMAGES, exist_ok=True)
all_images = os.listdir(FAIL_IMAGES)

augment = transforms.Compose([
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomRotation(degrees=30),
    transforms.ColorJitter(brightness=0.4, contrast=0.4),
    transforms.GaussianBlur(kernel_size=5),
])

aug_count = 0

while aug_count < TARGET:
    # Pick a random image
    filename = random.choice(all_images)
    img = Image.open(os.path.join(FAIL_IMAGES, filename)).convert('RGB')

    # Generate 3 augmented versions from it
    for i in range(3):
        new_img = augment(img)
        new_img.save(os.path.join(FAIL_IMAGES, f'aug_{aug_count}.jpg'))
        aug_count += 1

        if aug_count >= TARGET:
            break

print(f'Done! {aug_count} augmented images saved.')
```
